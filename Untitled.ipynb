```python
data = open('output.txt').read().split('\n')
n = len(data)


def parse(a, b, c, d):
    name = a.split(':')[0]
    t = float(b.split()[0])
    m = int(c.split()[-1])
    s = int(d.split()[-1])
    return name, t, m, s

xs = []

y_time_1 = []
y_time_2 = []
y_time_3 = []
y_time_4 = []

y_mult_1 = []
y_mult_2 = []
y_mult_3 = []
y_mult_4 = []

y_sum_1 = []
y_sum_2 = []
y_sum_3 = []
y_sum_4 = []


for i in range(0, n - 1, 25):
    current_size = data[i].split()[-1]
    xs.append(current_size)
    
    
    n1, t1, m1, s1 = parse(data[i + 2], data[i + 3], data[i + 4], data[i + 5])
    n2, t2, m2, s2 = parse(data[i + 7], data[i + 8], data[i + 9], data[i + 10])
    n3, t3, m3, s3 = parse(data[i + 13], data[i + 14], data[i + 15], data[i + 16])
    n4, t4, m4, s4 = parse(data[i + 19], data[i + 20], data[i + 21], data[i + 22])
    
    y_time_1.append(t1)
    y_time_2.append(t2)
    y_time_3.append(t3)
    y_time_4.append(t4)
    
    y_mult_1.append(m1)
    y_mult_2.append(m2)
    y_mult_3.append(m3)
    y_mult_4.append(m4)
    
    y_sum_1.append(s1)
    y_sum_2.append(s2)
    y_sum_3.append(s3)
    y_sum_4.append(s4)
    
    
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# 1 - simple
# 2 - fft
# 3 - fht
# 4 - dht
    
for (name, a, b, c, d) in [('time', y_time_1, y_time_2, y_time_3, y_time_4),
                     ('mult', y_mult_1, y_mult_2, y_mult_3, y_mult_4),
                     ('sum', y_sum_1, y_sum_2, y_sum_3, y_sum_4)]:
    
    max_y = max(max(v) for v in [a, b, c, d])
    min_y = min(min(v) for v in [a, b, c, d])
    data = {
            n1:a,
            n2:b,
            n3:c,
            n4:d
           }
    
    df = pd.DataFrame(data)
    plt.figure(figsize=(12,8))
    plt.title(name)
    plt.axis([0, len(xs), min_y - 5, max_y + 5])
    plt.plot(xs, df)
    plt.legend(data, loc=2)
    plt.show()
```
